```python
# Importing Libraries 
import pandas as pd
```

```python
# Reading CSV File in Data-Frame
df = pd.read_csv('responses.csv', low_memory= False)
df.head()
```

```python
# Converting Object Dtype to date_time
df.created_time = pd.to_datetime(df.created_time)
```

```python
# Extracting Year From Created_Time attribute
year = df['created_time'].map(lambda created_time: created_time.strftime("%Y"))
```

```python
# Appendind Year as 'year' in DataFrame
df.insert(0, "year", year, True)
```

```python
df['year'] = df['year'].astype('int')
```

```python
# Working on Year 2021-22
df_21_22 = df[df['year'] > 2020]
```

```python
# Grouping According To Requirement
gp = df_21_22.groupby(['year','brand_name','user_id','story_id','video_id']).node_key.agg(['count','max','min','unique'])
gp
```

```python
# Saving Grouped Data In Excel
gp.to_excel("result.xlsx")
```
